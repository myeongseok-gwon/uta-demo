```python
import os
import pandas as pd
import csv

# 대상 디렉토리 설정 (현재 디렉토리)
directory = '.'

# 디렉토리 내의 모든 파일 확인
for filename in os.listdir(directory):
    if filename.startswith("top_") and filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        print(f"Processing {filename}...")
        
        # CSV 파일 읽기
        df = pd.read_csv(filepath)
        
        # 'influencer' 컬럼이 있다면 이를 기준으로, 없으면 'name' 컬럼으로 중복 제거
        if "influencer" in df.columns:
            df_unique = df.drop_duplicates(subset=["influencer"], keep="first")
        elif "name" in df.columns:
            df_unique = df.drop_duplicates(subset=["name"], keep="first")
        else:
            print(f"  -> {filename}에는 'influencer' 또는 'name' 컬럼이 없습니다. 건너뜁니다.")
            continue
        
        # deduplicated CSV 파일 저장 (모든 필드를 큰따옴표로 감쌈)
        df_unique.to_csv(filepath, index=False, encoding="utf-8-sig", quoting=csv.QUOTE_ALL)
        print(f"  -> {filename} 중복 제거 완료 및 저장됨.")
```
